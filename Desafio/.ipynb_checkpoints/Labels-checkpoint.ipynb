```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import openfoodfacts
```

```python
labels = pd.read_json("https://world.openfoodfacts.org/labels.json")
```

```python
ids=[]
product_count=[]
for i in range(len(labels)):
    ids.append(labels['tags'][i]['id'])
    product_count.append(labels['tags'][i]['products'])

```

```python
df = pd.DataFrame(zip(ids, product_count), columns=['id','count'])
```

```python
df[['language','content']] = df['id'].str.split(':',expand=True)
```
